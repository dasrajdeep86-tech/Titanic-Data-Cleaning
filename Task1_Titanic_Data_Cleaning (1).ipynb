# Step 4: Handle Missing Values

Missing values can reduce the quality of analysis.

Actions performed:
- Replaced missing values in **Age** with the median.
- Replaced missing values in **Embarked** with the mode.
- Removed the **Cabin** column because it contains too many missing values.

### Observation

All important missing values have been handled successfully.

The dataset is now cleaner and more reliable for analysis.



```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.drop("Cabin", axis=1, inplace=True)
```

# Step 5: Remove Duplicate Records

Duplicate records may produce incorrect analysis results.

This step removes duplicate rows from the dataset.

### Observation

Duplicate records were checked and removed successfully.

This improves data quality.

```python
df.drop_duplicates(inplace=True)
```

# Step 6: Convert Data Types

Data types were converted to improve efficiency and readability.

```python
df.dtypes
```

```python
df["Survived"] = df["Survived"].astype(int)
df["Pclass"] = df["Pclass"].astype("category")
```

# Step 7: Rename Columns

Column names were renamed to make the dataset easier to understand and maintain.

```python
df.rename(columns={
    "PassengerId":"Passenger_ID",
    "Pclass":"Passenger_Class",
    "Sex":"Gender"
}, inplace=True)
```
